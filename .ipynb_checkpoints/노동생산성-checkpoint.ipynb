```python
# 전처리
labor= pd.read_excel('labor1.xls')
labor1 = pd.read_excel('labor2.xls')

labor
labor1
```

```python
columns = ['산업','2016','2017','2018','2019']
labor.columns=columns
labor1.columns=columns
labor
labor1
```

```python
labor['산업']=labor['산업'].str.replace(r"\(.*\)","")
labor1['산업']=labor1['산업'].str.replace(r"\(.*\)","")
```

```python
allabor = pd.concat([labor,labor1])
```

```python

allabor
```

```python
allabor=allabor.set_index('산업')
allabor = allabor.loc[~allabor.index.duplicated(keep='first')]
```

```python
plt.figure(figsize=(20,10))
bar_width = 0.18
index = np.arange(69)

plt.title('한국 주요 산업별 노동 생산성 2019')
plt.bar(index,allabor['2019'],bar_width)

plt.xticks(index, industry, rotation=90)
plt.xlabel('산업')
plt.ylabel('노동 생산성')
plt.show();
```
